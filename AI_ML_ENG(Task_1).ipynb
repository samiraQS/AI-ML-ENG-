# Task 01: Exploring and Visualizing a Simple Dataset


```python
import pandas as pd  # IMPORTING Pandas TO LOAD DATASET
import seaborn as sns # IMPORTING Seaborn for statistical data visualizations
import matplotlib.pyplot as plt # Matplotlib for plotting graphs and charts
```

```python
iris = sns.load_dataset('iris') # Load the built-in Iris dataset from the Seaborn library
print(iris.head()) # Print first 5 rows
```

```python
print("Shape of datasets:") # Display the dimensions (rows and columns) of the dataset.
print(iris.shape)
```

```python
print("Column Names:") # For coloumns names in the dataset
print(iris.columns)
```

```python
print("First ten rows:")
print(iris.head(10)) # To display first 10 rows in the dataset
```

```python
print("Dataset information:")
print(iris.info()) # To display information about the dataset, such as column names, data types, non-null values, and memory usage
```

```python
print("Dataset summary:")
print(iris.describe()) # SUMMARY STATISTICS OF NUMERICAL FEATURES
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot( # scatter plot visualization relationship between sepal length and sepal width for different iris species
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=iris
)

plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
iris.hist(figsize=(8,8)) ## Create histograms for all numerical features in the Iris dataset the figure size is set to 10 inches wide and 8 inches high
plt.suptitle('Histogram')
plt.show()
```

```python
plt.figure(figsize=(7,6))

sns.boxplot(data=iris.iloc[:, :-1]) # Create box plots for all numerical features in the Iris dataset iris.iloc[:,:-1] selects all feature columns except the last column (species)

plt.title('Box Plot of Iris Features')
plt.show()
```
